```python
corr = sales.select_dtypes(include='number').corr(method='pearson') # кольоряція
print(corr)
```

```python
fig = plt.figure(figsize=(8,8))
plt.matshow(corr, cmap='RdBu', fignum=fig.number)

#підписи для рядків та стовпчиків матриці
plt.xticks(range(len(corr.columns)), corr.columns, rotation='vertical')
plt.yticks(range(len(corr.columns)), corr.columns)

# кольорова гама
plt.colorbar()
plt.clim(-1,1)
```

```python
df = pd.DataFrame({'x':[1, 2, 3, 4, 5, 6, 7, 8, 9], # ДЛЯ ВИЯВЛЕННЯ будьякої зростаючої фбо спадаючої залежності є коефіцієнт кореляції Спірмена
                   'y':[1, 2, 23, 24, 105, 106, 507, 508, 10_009]})
print(f"pearson:  {df.corr(method='pearson').loc['x', 'y']}")
print()
print(f"speaman: {df.corr(method='spearman').loc['x', 'y']}")
```

```python
corr = sales.select_dtypes(include='number').corr(method='spearman')

fig = plt.figure(figsize=(8,8))
plt.matshow(corr, cmap='RdBu', fignum=fig.number)
plt.xticks(range(len(corr.columns)), corr.columns, rotation='vertical')
plt.yticks(range(len(corr.columns)), corr.columns)
plt.colorbar()
plt.clim(-1, 1)
```
